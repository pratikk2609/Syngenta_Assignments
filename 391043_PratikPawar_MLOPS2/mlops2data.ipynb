```python
import pandas as pd

class ChunkIterator:
    def __init__(self, file_path, chunk_size):
        self.file_path = file_path
        self.chunk_size = chunk_size
        self.chunk_generator = pd.read_csv(file_path, chunksize=chunk_size)
    
    def __iter__(self):
        return self
    
    def __next__(self):
        try:
            chunk = next(self.chunk_generator)
            return chunk
        except StopIteration:
            raise StopIteration
    
    def calculate_statistics(self, chunk):
        statistics = {
            'mean': chunk.mean(numeric_only=True).to_dict(),
            'std': chunk.std(numeric_only=True).to_dict(),
            'min': chunk.min(numeric_only=True).to_dict(),
            'max': chunk.max(numeric_only=True).to_dict(),
            'count': chunk.count(numeric_only=True).to_dict()
        }
        return statistics
```

```python
file_path = 'mlops2data.csv'
chunk_size = 1000

iterator = ChunkIterator(file_path, chunk_size)

for chunk in iterator:
    stats = iterator.calculate_statistics(chunk)
    print("Chunk Statistics:")
    print("Mean:", stats['mean'])
    print("Std Dev:", stats['std'])
    print("Min:", stats['min'])
    print("Max:", stats['max'])
    print("Count:", stats['count'])
    print("----------------------------------------------------")
```
